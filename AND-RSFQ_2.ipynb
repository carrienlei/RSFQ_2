```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("AND_result.csv")

drop_col = ['cus', 'real_flux', 'Pulse width info', 'Pulse magnitude info', 'Close to previous clock (0: no pulse, a: equal, b: close to later clock, c: close to prev clock)']
df = df.drop(drop_col, axis=1)
df = df.dropna()


for index, row in df.iterrows():
    val = str(row['config']).split('_')
    df.at[index, 'INPUT_1'] = val[0]
    df.at[index, 'Spl Num 1'] = val[2]

    df.at[index, 'INPUT_2'] = val[3]
    df.at[index, 'Spl Num 2'] = val[5]

    df.at[index, 'LOAD'] = val[9]
    df.at[index, 'Spl Num Load'] = val[8]

    real_output = eval(str(row['real_output']))
    df.at[index, 'Real Output'] = real_output['n200']

    gold_output = eval(str(row['gold_output']))
    df.at[index, 'Gold Output'] = gold_output['n200']
    
    pv = str(row['MC']).split(',')
    df.at[index, 'Process Variation'] = pv[1]

drop_col = ['MC', 'real_output', 'gold_output']
df = df.drop(drop_col, axis=1)

df.head()
```

```python
pv_01 = df[df['Process Variation'] == "0.1"]
pv_01.head()

grouped = pv_01.groupby('config')

for config, group in grouped:
    print(f'Config Value: {config}:')
    for index, row in group.iterrows():
        error0 = "00,00"
        count0 = 0

        error1 = "00,01"
        count1 = 0

        error2 = "00,10"
        count2 = 0

        error3 = "00,11"
        count3 = 0

        error4 = "01,00"
        count4 = 0
        
        error5 = "01,01"
        count5 = 0
        
        error6 = "01,10"
        count6 = 0
        
        error7 = "01,11"
        count7 = 0
        
        error8 = "10,00"
        count8 = 0
        
        error9 = "10,01"
        count9 = 0
        
        error10 = "10,10"
        count10 = 0
        
        error11 = "10,11"
        count11 = 0
        
        error12 = "11,00"
        count12 = 0
        
        error13 = "11,01"
        count13 = 0
        
        error14 = "11,10"
        count14 = 0
        
        error15 = "11,11"
        count15 = 0
        

        bit1 = row['Real Output']
        bit2 = row['Gold Output']
        
        differing_positions = []
        for i in range(len(bit1)):
            if bit1[i] != bit2[i]:
                differing_positions.append(i)

        patterns = row['pattern'].split(',')
        differing_elements = [patterns[i-2] +"," + patterns[i-1] for i in differing_positions]

        for val in differing_elements:
            if val == error0:
                count0+=1
            elif val == error1:
                count1 +=1
            elif val == error2:
                count2 +=1
            elif val == error3:
                count3 +=1
            elif val == error4:
                count4 +=1
            elif val == error5:
                count5 +=1
            elif val == error6:
                count6 +=1
            elif val == error7:
                count7 +=1
            elif val == error8:
                count8 +=1
            elif val == error9:
                count9 +=1
            elif val == error10:
                count10 +=1
            elif val == error11:
                count11 +=1
            elif val == error12:
                count12 +=1
            elif val == error13:
                count13 +=1
            elif val == error14:
                count14 +=1
            else:
                count15 +=1

        group.at[index, 'Number of Errors'] = len(differing_elements)
        group.at[index, 'Prior Two Input Patterns to Error'] = str(differing_elements)   
        group.at[index,"Unique Error Sequences: "] = len(set(differing_elements))


    print(str(differing_elements))
    print ("Error 0 Probability: " , (int(count0)/group.shape[0]))
    print ("Error 1 Probability: " , (int(count1)/group.shape[0]))
    print ("Error 2 Probability: " , (int(count2)/group.shape[0]))
    print ("Error 3 Probability: " , (int(count3)/group.shape[0]))
    print ("Error 4 Probability: " , (int(count4)/group.shape[0]))
    print ("Error 5 Probability: " , (int(count5)/group.shape[0]))
    print ("Error 6 Probability: " , (int(count6)/group.shape[0]))
    print ("Error 7 Probability: " , (int(count7)/group.shape[0]))
    print ("Error 8 Probability: " , (int(count8)/group.shape[0]))
    print ("Error 9 Probability: " , (int(count9)/group.shape[0]))
    print ("Error 10 Probability: " , (int(count10)/group.shape[0]))
    print ("Error 11 Probability: " , (int(count11)/group.shape[0]))
    print ("Error 12 Probability: " , (int(count12)/group.shape[0]))
    print ("Error 13 Probability: " , (int(count13)/group.shape[0]))
    print ("Error 14 Probability: " , (int(count14)/group.shape[0]))
    print ("Error 15 Probability: " , (int(count15)/group.shape[0]))
   

    display(group)
    print('\n')
```
